#Import Library

```python
# Import Library
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

#Get Link From JobStreet

```python
page = 1
url = f"https://id.jobstreet.com/id/data-jobs?page={page}"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

print(response)
```

#Get HTML Element

```python
job_listings = soup.find_all('div', class_="_8i9ct20 nbwpkd6m") # Get div element
jobs = []

for job in job_listings:
  article = job.find('article')
  if article != None:
    print(article)
    jobs.append(article)
  span = job.find('span')
  if span != None:
    print(span)
    jobs.append(span)
```

```python
list_clean_text = []
for job in jobs:
  clean_text = job.get_text(strip=True)
  list_clean_text.append(clean_text)
  print(clean_text)
```

```python
data = pd.DataFrame(columns=['job_name', 'company_name', 'job_type', 'job_category', 'job_sub_category', 'sallary'])
for job in list_clean_text:
  match = re.search(  r'ago(.*?)di',job)
  if match != None:
    job_name = match.group(1)
    match = re.search(  r'di(.*?)Ini',job)
    if match != None:
      company_name = match.group(1)
    match = re.search(  r'kerja (.*?) ',job)
    if match != None:
      job_type = match.group(1)
      if "Kontrak" in job_type:
          job_type = "Kontrak"
      else:
          job_type = match.group(1)
    match = re.search(  r'subClassification: (.*?)classification:',job)
    if match != None:
      job_category = match.group(1)
    match = re.search(  r'classification: (.*?) ',job)
    if match != None:
      job_sub_category = match.group(1)
    match = re.search(r'Rp\s*([\d\.]+)\s*\s*Rp\s*([\d\.]+)', job)
    if match != None:
      sallary = f"{match.group(1)}, {match.group(2)}"
    else:
      sallary = None
    data = pd.concat([data, pd.DataFrame([{'job_name': job_name, 'company_name': company_name, 'job_type': job_type, 'job_category': job_category, 'job_sub_category': job_sub_category, 'sallary': sallary}])], ignore_index=True)
```

#Crawling Web Page

```python
page = 1
url = f"https://id.jobstreet.com/id/data-jobs?page={page}"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
jobs_crawling = []

for page in range(1, 11):
    url = f"https://id.jobstreet.com/id/data-jobs?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    job_listings = soup.find_all('div', class_="_8i9ct20 nbwpkd6m")
    for job in job_listings:
      article = job.find('article')
      if article != None:
        jobs_crawling.append(article)
      span = job.find('span')
      if span != None:
        jobs_crawling.append(span)
    print(f'page ke-{page} telah berhasil di scrap')
```

```python
list_job_id = []
for job in jobs_crawling: # get job
    if job.name == 'article':
        job_id = job['data-job-id']
        list_job_id.append(job_id)
```

```python
print(len(list_job_id))
```

```python
list_clean_text_crawling = []
for job in jobs_crawling:
  clean_text = job.get_text(strip=True)
  list_clean_text_crawling.append(clean_text)
```

```python
len(list_clean_text_crawling)
```

```python
data_crawling = pd.DataFrame(columns=['job_name', 'company_name', 'job_type', 'job_category', 'job_sub_category', 'sallary'])
for job in list_clean_text_crawling:
  match = re.search(  r'ago(.*?)di',job)
  if match != None:
    job_name = match.group(1)
    match = re.search(  r'di(.*?)Ini',job)
    if match != None:
      company_name = match.group(1)
    match = re.search(  r'kerja (.*?) ',job)
    if match != None:
      job_type = match.group(1)
      if "Kontrak" in job_type:
          job_type = "Kontrak"
      else:
          job_type = match.group(1)
    match = re.search(  r'subClassification: (.*?)classification:',job)
    if match != None:
      job_category = match.group(1)
    match = re.search(  r'classification: (.*?) ',job)
    if match != None:
      job_sub_category = match.group(1)
    match = re.search(r'Rp\s*([\d\.]+)\s*\s*Rp\s*([\d\.]+)', job)
    if match != None:
      sallary = f"{match.group(1)}, {match.group(2)}"
    else:
      sallary = None
    data_crawling = pd.concat([data_crawling, pd.DataFrame([{'job_name': job_name, 'company_name': company_name, 'job_type': job_type, 'job_category': job_category, 'job_sub_category': job_sub_category, 'sallary': sallary}])], ignore_index=True)
```

```python
data_crawling["job_id"] = list_job_id
```

# More Detail Scrape

```python
count = 0
job_desc = []
for id in data_crawling["job_id"]:
  count += 1
  url = f"https://id.jobstreet.com/id/job/{id}"
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  detail = soup.find_all('div', class_="_8i9ct20 tt829s0")
  job_desc.append(detail)
  print(f'job - {count} telah berhasil di scrap')
```

```python
desc_job = []
for desc in job_desc:
  for x in desc:
    clean = x.get_text(strip=True)
    desc_job.append(clean)
```

```python
data_crawling["job_desc"] = desc_job
```
